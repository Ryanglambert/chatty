# Random Forest

```python
ACT_MASK = train['act'].notnull()
# X = train[ACT_MASK].values
X = vocab[ACT_MASK]
y = train['act'][ACT_MASK].values

clf = Pipeline([
    ('countvectorizer', CountVectorizer(analyzer=tokens.token_splitter)),
    ('random', RandomOverSampler('minority')),
    ('RandomForest', RandomForestClassifier(n_jobs=-1, verbose=True))
])
```

```python
results = data.cv_stratified_shuffle(X, y, clf, splits=2)
```

```python
plot.plot_prec_rec(results, 'RandomForest', normalize_confusion_matrix=True)
```

```python
names = results['models'][0].named_steps['countvectorizer'].get_feature_names()

importances = results['models'][0].named_steps['RandomForest'].feature_importances_
list(filter(lambda x: x[0] < 50, enumerate(sorted(zip(names, importances), key=lambda x: x[1], reverse=True))))
```

```python
plt.plot(sorted(importances, reverse=True))
plt.xlim(0, 2000)
plt.ylim(0, .05)
```

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.ensemble import VotingClassifier, RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.calibration import CalibratedClassifierCV # <<<<< use to finalize
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
log_reg_pca = Pipeline([
    ('pca', TruncatedSVD(n_components=30)),
    ('logreg', LogisticRegression(n_jobs=-1, solver='newton-cg'))
])

vt = VotingClassifier([
    ('logreg', LogisticRegression(n_jobs=-1, solver='newton-cg')),
    ('svc', SGDClassifier(loss='log', n_jobs=-1)),
    ('adaboost', AdaBoostClassifier())
], n_jobs=1, voting='soft')

ACT_MASK = train['act'].notnull()
# X = train[ACT_MASK].values
X = vocab[ACT_MASK]
y = train['act'][ACT_MASK].values

clf = Pipeline([
#     ('countvectorizer', CountVectorizer(analyzer=tokens.token_splitter)),
    ('tfidf_vec', TfidfVectorizer(analyzer=tokens.token_splitter, use_idf=False)),
    ('random', RandomOverSampler('minority')),
    ('votingclassifier', vt)
])
```

```python
results = data.cv_stratified_shuffle(X, y, clf, splits=2)
```

# Try with utterances t-1 & t-2 include POS bigrams of the document

```python
from sklearn.pipeline import FeatureUnion

from sklearn.decomposition import TruncatedSVD

from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import RFE, RFECV
from sklearn.cross_validation import StratifiedKFold
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.calibration import CalibratedClassifierCV
from xgboost import XGBClassifier

from chatty.utils.vectorize import CountVectorizerColumnN, ColumnGetter, Cosiner
```

```python

%reload_ext chatty.utils.sklearn_fix
```

```python
tokens.list_vocabs()
```

```python
vocab = tokens.load_vocab('chunk_pos_bigram', 'lemma_ngram_2', 'pos_ngram_2', 'lemma_ngram_3', 'pos_ngram_3', 'word')['vocab']
vocab = [tokens.token_joiner(i) for i in vocab]
vocab = np.array(vocab)

train['vocab'] = vocab

VECTOR_MASK = []
for i in range(300):
    field = 'vector_{}'.format(i)
    VECTOR_MASK.append(field)
    train[field] = train_vecs[:, i]
```

```python
LAG_MASK = ['vocab_t-2', 'vocab_t-1']
train[LAG_MASK] = data.get_lags(train[['vocab']], lag_range=[2, 1])
VECTOR_LAG_MASK = [i + '_t-1' for i in VECTOR_MASK] + [i + '_t-2' for i in VECTOR_MASK]
train[VECTOR_LAG_MASK] = data.get_lags(train[VECTOR_MASK], lag_range=[1, 2])
ACT_MASK = train['act'].notnull() \
         & train['vocab'].notnull() \
         & train['vocab_t-1'].notnull() \
         & train[VECTOR_LAG_MASK].notnull().any(axis=1) \
         & train['vocab_t-2'].notnull()
```

```python
# logregpca = Pipeline([
#     ('pca', TruncatedSVD(n_components=1000)),
#     ('logreg', CalibratedClassifierCV(SGDClassifier(loss='log', max_iter=1000,
#                                                     tol=1e-3, n_jobs=-1, verbose=True)))
# ])

# svc = SGDClassifier(loss='hinge', n_jobs=-1, max_iter=1000, tol=1e-3, verbose=True)
logreg = CalibratedClassifierCV(SGDClassifier(loss='log',
                                              n_jobs=-1,
                                              max_iter=1000,
                                              tol=1e-3,
                                              verbose=True,
                                              penalty='l1',
                                              learning_rate=1e-5))
# adaboost = AdaBoostClassifier()
# gbm = GradientBoostingClassifier()

xgb = XGBClassifier(max_depth=3, n_jobs=-1, silent=False)
mlp = MLPClassifier(hidden_layer_sizes=(5,), max_iter=50) # so slow

vt = VotingClassifier([
    ('logreg', logreg),
#     ('logregpca', logregpca),
    ('xgb', xgb)
#     ('mlp', mlp),
#     ('adaboost', adaboost)
], n_jobs=1, voting='soft')


fu = FeatureUnion([('vocab', CountVectorizerColumnN(analyzer=tokens.token_splitter,
                                                        colnum=1, normalize=True)),
                   ('vocab_t-1', CountVectorizerColumnN(analyzer=tokens.token_splitter,
                                                    colnum=0, normalize=True)),
                   ('vectors', ColumnGetter(colnum_range=(2, 302))),
                   ('vecs_t0t1', Cosiner([(2, 302), (302, 602)])),
                   ('vecs_t0t2', Cosiner([(2, 302), (602, 902)]))])
FEATURE_MASK = ['vocab_t-1', 'vocab'] + VECTOR_MASK + VECTOR_LAG_MASK
X = train[FEATURE_MASK][ACT_MASK].values
y = train['act'][ACT_MASK].values

clf = Pipeline([
#     ('countvectorizer', CountVectorizer(analyzer=tokens.token_splitter)),
    ('feature_union_countvecs', fu),
#     ('tfidf_vec', TfidfVectorizer(analyzer=tokens.token_splitter, use_idf=False)),
    ('random', RandomOverSampler('minority')),
    ('votingclassifier', vt)
#     ('logreg', logreg)
#     ('xgb', xgb)
])
```

```python
results = data.cv_stratified_shuffle(X, y, clf, splits=2, probability=True)
```
